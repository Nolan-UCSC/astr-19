import matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

Select your favorite distribution (other than uniform) and pull 1000 random numbers from that distribution.

```python
# Mean and standard deviation for the normal distribution
mu, sigma = 0, 0.1

# Drawing 1000 samples from the normal distribution
normal_samples = np.random.normal(mu, sigma, 1000)
```

Histogram the random numbers into 100 bins, and plot the histogram. Label your axes and save the figure as a PDF.

```python
# Create the histogram for the normal distribution samples
plt.hist(normal_samples, bins=100, alpha=0.75)

# Label the axes
plt.xlabel('Value')
plt.ylabel('Frequency')

# Add a title
plt.title('Normal Distribution Histogram')

# Save the figure
plt.savefig('normal_distribution_histogram.pdf')

# Show the plot
plt.show()
```
