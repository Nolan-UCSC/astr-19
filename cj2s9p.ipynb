import matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

Use numpy to pull 1000 random numbers distributed uniformly between [0,1]

```python
random_numbers = np.random.uniform(0, 1, 1000)
```

Histogram the random numbers into 100 bins, and plot the histogram. Label your axes and save the figure as a PDF.

```python
# Create a histogram of the random numbers
plt.hist(random_numbers, bins=100, alpha=0.7)

# Label the axes
plt.xlabel('Value')
plt.ylabel('Frequency')

# Add a title
plt.title('Histogram')

# Save the figure
plt.savefig('uniformhistogram.pdf')

# Show the plot
plt.show()
```
